```python
import json
from pathlib import Path
import sys
from functools import partial
sys.path.append("../") 
from src.models.llm.llama import LlamaLLM, EVAL_SYSTEM_INSTRUCTIONS
from src.data.utils import read_jsonl
```

```python
eval_llm = LlamaLLM(prompt=EVAL_SYSTEM_INSTRUCTIONS)
```

```python
with open("results.json") as f:
    results = json.load(f)

json_objs = read_jsonl("../data/raw/task1/crag_task_1_v2.jsonl", -1)
json_objs = [json_obj for json_obj in json_objs if json_obj["split"] == 0]

misses = []
for i, result in enumerate(results):
    json_obj = json_objs[i]
    for k, v in result.items():
        query = f"Is '{k}' equivalent to '{v}'?"
        is_correct = eval_llm.process_example(query)
        misses.append(is_correct)
```

```python
for i, result in enumerate(results):
    is_correct = int(misses[i]) == 1
    for k, v in result.items():
        print(f"i: {i}\nis_correct: {is_correct}\n:gold answer: {k}\npred answer: {v}")
```
